## 업데이트 순 웹툰 추출

### selenium 사용

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver_path = 'C:\\Users\\qkrwn\\JupyterLab\\JupyterLa\\JupyterLab\\chromedriver_win32\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)

driver.get("https://comic.naver.com/webtoon/weekday?order=User")



update = '#content > div.view_type > ul > li:nth-child(2) > a > span'
driver.find_element_by_css_selector(update).click()

time.sleep(3)

html = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(html)
webtoons = soup.select('.col_inner')
title = []

for webtoon in webtoons:
    for i in webtoon.find_all('a', class_='title'):
        title.append(i.get_text())
    
title
```

## 요일별 추천 웹툰 이름, 별점 추출

```python
driver_path = 'C:\\Users\\qkrwn\\JupyterLab\\JupyterLa\\JupyterLab\\chromedriver_win32\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)

driver.get("https://comic.naver.com/webtoon/weekday?order=User")

mon = '#content > ul > li:nth-child(2) > a'
driver.find_element_by_css_selector(mon).click()
time.sleep(3)
monday = driver.page_source

tue = '#content > ul > li:nth-child(3) > a'
driver.find_element_by_css_selector(tue).click()
time.sleep(3)
tuesday = driver.page_source

wed = '#content > ul > li:nth-child(4) > a'
driver.find_element_by_css_selector(wed).click()
time.sleep(3)
wednesday = driver.page_source

thu = '#content > ul > li:nth-child(5) > a'
driver.find_element_by_css_selector(thu).click()
time.sleep(3)
thursday = driver.page_source

fri = '#content > ul > li:nth-child(6) > a'
driver.find_element_by_css_selector(fri).click()
time.sleep(3)
friday = driver.page_source

sat = '#content > ul > li:nth-child(7) > a'
driver.find_element_by_css_selector(sat).click()
time.sleep(3)
saturday = driver.page_source

sun = '#content > ul > li:nth-child(8) > a'
driver.find_element_by_css_selector(sun).click()
time.sleep(3)
sunday = driver.page_source
driver.close()
```

```python
mon_title = []
tue_title = []
wed_title = []
thu_title = []
fri_title = []
sat_title = []
sun_title = []
week_title = []

soup = BeautifulSoup(monday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        mon_title.append(i.get_text())
soup = BeautifulSoup(tuesday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        tue_title.append(i.get_text())
soup = BeautifulSoup(wednesday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        wed_title.append(i.get_text())
soup = BeautifulSoup(thursday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        thu_title.append(i.get_text())
soup = BeautifulSoup(friday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        fri_title.append(i.get_text())
soup = BeautifulSoup(saturday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        sat_title.append(i.get_text())
soup = BeautifulSoup(sunday)
webtoons = soup.select('.webtoon_spot')
for webtoon in webtoons:
    for i in webtoon.find_all('strong'):
        sun_title.append(i.get_text())


week_title.append(mon_title)
week_title.append(tue_title)
week_title.append(wed_title)
week_title.append(thu_title)
week_title.append(fri_title)
week_title.append(sat_title)
week_title.append(sun_title)

print(week_title)
```

```python
indexs = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
columns = ['웹툰이름','별점', '웹툰이름','별점', '웹툰이름','별점']
df = pd.DataFrame(week_title,columns=columns, index = indexs)
df
```
