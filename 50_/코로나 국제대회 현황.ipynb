```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('corona_korea.xlsx')
df
```

```python
df.drop(['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'], axis = 1, inplace=True)
df.columns = ['일자', '계', '국내발생(명)', '해외유입(명)', '사망(명)']
df.drop([0, 1, 2, 3, 4], axis = 0, inplace=True)
df
```

```python
plt.figure(figsize=(10,5))
#sns.lineplot(x=df['일자'], y=df['계'],data=df['계'])
plt.plot(df['일자'], df['계'])
```

```python
df = pd.read_csv('corona_total.csv')
df
```

```python
df.columns = ['인덱스', '일자', '확진자수']
df
```

```python
df_1 = df[(df.인덱스 >= 8) & (df.인덱스 <=16)][['일자', '확진자수']]
df_1
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='일자', y='확진자수', data=df_1)
```
