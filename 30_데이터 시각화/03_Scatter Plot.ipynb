# 3. Scatter plot
* 좌표상의 점들을 표시하여 두 변수간의 관계를 나타내는 그래프

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50) # 0 
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
x2 = np.random.rand(50)
y2=np.random.rand(50)
plt.scatter(x2,y2)
plt.show()
```

```python
# Color 지정하기
plt.scatter(x,y, color='r')
plt.show()
```

```python
# 점마다 다른 Color 지정하기
y=[1,2,3]
x= [1,2,3]
colors=['r','g','b']
plt.scatter(x,y, c=colors)
plt.show()
```

```python
# ColorMap(색상지도) 사용하기
# 데이터값을 기반으로 색상 표현하기
plt.colormaps()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,c=y, cmap='Blues')
```

```python
# Marker Size 변경하기
size = np.random.randint(50) * 100
plt.scatter(x,y,c=x,cmap='Blues', s=size)
```

```python
# 투명도 (alpha값, 0~1)
plt.scatter(x,y,c=x, cmap='Blues', s=size, alpha=0.5)
```
