# 4. Bar Plot
* 막대 그래프로 여러 값들을 비교할 때 유용함
* 데이터
    * 수치형 데이터/양적 데이터(numerical data) : 키, 몸무게, 시력, 머리카락 갯수
    * 범주형 데이터(Categorical data) : 성별, 혈액형, 출신 지역, 모의고사 등급
* x축에는 범주형 데이터 y 축에는 수치형 데이터 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 년도별 출생자수 비교
years = ['2018','2019','2020']
number = [100, 90, 80]

plt.bar(years, number, width = 776)
plt.title('년도별 출생자수',size = 20)
plt.xlabel('년도')
plt.ylabel('출생자수')
plt.show()
```

```python
# 수평 막대그래프 그리기
# barh(y,width)
plt.barh(years, number, height=0.5)
plt.title('년도별 출생자수',size = 20)
plt.xlabel('출생자수')
plt.ylabel('년도')
plt.show()
```

```python
# 년도, 성별로 출생자수비교하기
male = [100, 90, 80]
female = [90, 80, 70]
x = np.arange(3) # 0~2

plt.bar(x, male, width=0.5)
plt.bar(x,female, width=0.5)
plt.xticks(x,years)
plt.show()
```

```python
bar_width = 0.25
plt.bar(x - bar_width/2, male, width=bar_width)
plt.bar(x + bar_width/2, female, width=bar_width)
plt.legend(['male', 'female'])
plt.xticks(x, years)
```
